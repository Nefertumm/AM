## Perceptr贸n Simple


Un Perceptr贸n simple (PS) es la red neuronal m谩s sencilla que se puede considerar, est谩 conformado por una sola neurona que posee N entradas y una funci贸n de transferencia de tipo umbral, tal como se ve en la siguiente figura:

Haciendo un peque帽o cambio y llevando el umbral $\theta$ hacia la izquierda en las ecuaciones anteriores, podemos definir un nuevo peso
$w_0 = -\theta$ y la entrada $x_0 = 1$ para poder escribir la salida de forma m谩s compacta:

<pre>

Podemos representar la salida con la siguiente expresi贸n: $$y = sign(\sum\limits_{i=0}^N {x_i w_i})$$

_**sign**_ corresponde a la funci贸n signo y es la _funci贸n de activaci贸n_ del perceptr贸n simple (tambi茅n podr铆a ser la funci贸n heaviside). M谩s adelante veremos que existen otras funciones de activaci贸n para otras aplicaciones.

El PS permite resolver problemas linealmente separables mediante una recta o un hiperplano de separaci贸n con ordenada al origen distinta de cero gracias al t茅rmino de _Bias_

A continuaci贸n tenemos el algoritmo de entrenamiento o _Regla del Perceptr贸n Simple_ 

Esta regla puede implementarse siguiendo estos pasos:
1. Inicializar el vector de pesos w con valores aleatorios entre 0 y 1.
2. Presentar un patr贸n de entrada x y calcular la salida $$y = sign(x_0 w_0 + \sum\limits_{i=1}^N {x_i w_i})$$  Recordemos que $w_0$ es el t茅rmino de bias y $x_0=1$, podemos representar la suma de productos usando un producto punto entre vectores: $$y = sign( w_0 + \vec{w}^T \cdot \vec{x})$$

3. Calcular el error entre la salida obtenida y la salida deseada $y_d$ $$e = y - y_d$$
4. Ajustar los pesos de la red con la siguiente ecuaci贸n: $$ \vec{w} = \vec{w} + \mu \vec{e} \cdot \vec{x}$$ $\mu$ es el coeficiente de aprendizaje o factor de entrenamiento (eta)
5. Volver al paso 2 y repetir el proceso

```python
# Librer铆as a importar
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from modules.utiles import plot_decision_regions
from sklearn.preprocessing import StandardScaler
```

## Dataset Iris
El conjunto de datos flor Iris contiene 50 muestras de cada una de tres especies de Iris (Iris setosa, Iris versicolor e Iris virginica), en total 150 muestras. Se midi贸 cuatro rasgos de cada muestra: el largo y ancho del s茅palo y p茅talo, en cent铆metros.

>Nosotros vamos a utilizar nuestro perceptr贸n para separar dos clases de flores (Iris setosa e Iris versicolor) seg煤n dos de sus caracter铆sticas: Largo de s茅palo (primer columna en el dataset) y largo de p茅talo (tercer columna).  

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
df.head()
```

# Graficamos el Dataset

```python
print(df.groupby(4).size())
#Seleccionamos solo las clases setosa y versicolor
#Y las columnas 1 y 3
X = df.iloc[0:100, [0, 2]].values
y = df.iloc[0:100, 4].values
y = np.where(y == 'Iris-setosa',0,1)
#Creamos un nuevo dataframe
```

```python
#Graficamos los datos
plt.scatter(X[:50, 0], X[:50, 1], color='red', marker='o', label='setosa')
plt.scatter(X[50:100, 0], X[50:100, 1], color='blue', marker='X', label='versicolor')
```

# Entrenamiento del Perceptr贸n

Separar los datos en conjunto de entrenamiento y prueba (20%)

```python
#Separamos los datos 80% entrenamiento y 20% prueba, seg煤n `y` y con shuffle
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y, shuffle = True)
```

```python
y_train
```

```python
#Escalamos los datos
scaler = StandardScaler()
scaler.fit(X_train)
X_train_std = scaler.transform(X_train)
X_test_std = scaler.transform(X_test)
```

Entrenamos el perceptr贸n con un `eta=0.1`, una semilla `random_state=1` y `shuffle=True`. Ver [documentaci贸n](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Perceptron.html)

```python
perceptron = Perceptron(eta0 = 0.1, random_state = 1, shuffle = True)
perceptron.fit(X_train_std, y_train)
```

**Coeficientes o pesos de la recta de separaci贸n**

```python
print(perceptron.coef_, perceptron.intercept_)
```

**Constante u ordenada al origen de la recta**

```python
#Obtenemos la pendiente y ordenada de la recta
m = -(perceptron.coef_[0][0] / perceptron.coef_[0][1])
b = -(perceptron.intercept_ / perceptron.coef_[0][1])
```

```python
print(m,b)
```

Recordemos la ecuaci贸n: $$y = sign(w_0 + \sum\limits_{i=1}^N {x_i w_i})$$

Para este problema con dos caracter铆sticas (x1 = longitudes del s茅palo y x2= longitudes de p茅talo ), la ecuaci贸n resulta:

$$y = sign(x_1 w_1 + x_2 w_2 + w_0)$$

Donde se separan las dos clases de flores, tendremos la frontera de decisi贸n, dada por la ecuaci贸n:

$$x_1 w_1 + x_2 w_2 + w_0 = 0$$

De esta ecuaci贸n podemos despejar la recta $x_2$ en funci贸n de $x_1$ que separa las clases en el espacio de soluciones

$$x_2 + x_1 \frac{w_1}{w_2} + \frac{w_0}{w_2} = 0$$

$$x_2 =  -\frac{w_1}{w_2}x_1 - \frac{w_0}{w_2} $$

La pendiente de la recta  $ m = -\frac{w_1}{w_2}$ y la ordenada al origen $ b = - \frac{w_0}{w_2}$

### _Ahora vamos a graficar esta recta!!_

Graficar la recta de separaci贸n del perceptr贸n simple. Graficar los datos de entrenamiento junto con los datos de prueba, muestre estos 煤ltimos con un borde negro para diferenciarlos de los datos de entrenamiento.

```python
#------------------ Representaci贸n de la recta ------------------------------------
#Graficamos el conjunto de entrenamiento
plt.scatter(X_train_std[:, 0][y_train == 0], X_train_std[:, 1][y_train == 0], color='red', marker='o', label='setosa')
plt.scatter(X_train_std[:, 0][y_train == 1], X_train_std[:, 1][y_train == 1], color='blue', marker='X', label='versicolor')

#Agregamos la recta
x = np.linspace(-2, 2, 100)
plt.plot(x, m*x + b, label='Recta', color = 'black')
plt.legend(loc='upper left')

# ----------------------------------------------------------------------------------
#Agregamos los datos de testeo
plt.scatter(X_test_std[:, 0][y_test == 0], X_test_std[:, 1][y_test == 0], edgecolors='black', color='red', marker='o', label='setosa')
plt.scatter(X_test_std[:, 0][y_test == 1], X_test_std[:, 1][y_test == 1], edgecolors='black', color='blue', marker='X', label='versicolor')
```

```python
#Concatenamos los datos escalados de prueba y testeo
X_std = np.vstack((X_train_std, X_test_std))
y_general = np.hstack((y_train, y_test))
#Obtenemos los indices de los datos de testeo
y_general
```

Ahora grafique los datos usando la funci贸n `plot_decision_regions` del m贸dulo `utiles.py`, grafique datos de entrenamiento y prueba (se le deben pasar los 铆ndices de los datos de prueba a la funci贸n).

```python
plot_decision_regions(X_std, y_general, perceptron, test_idx=range(80, 100))
```

Utilice la funci贸n `accuracy_score` para obtener el desempe帽o con los datos de prueba. [Documentaci贸n](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html)

```python
#Calculamos el score del clasificador
y_pred = perceptron.predict(X_test_std)
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```
